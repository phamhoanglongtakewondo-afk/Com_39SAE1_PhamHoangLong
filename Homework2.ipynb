```python
#Bai 3
a,b,c = map(float, input().split())
if a+b >c and a+c>b and b+c>a:
    print("La 3 canh cua tam giac")
    if a==b==c:
        print("La tam giac deu")
    elif a==b or a ==c or b==c:
        print("La tam giac can")
    elif a**2+b**2==c**2 or a**2+c**2== b**2 or b**2+c**2== a**2:
        print("La tam giac vuong")
    else:
        print("La tam giac binh thuong")
    import math
    p = (a+b+c)/2
    S_square = p*(p-a)*(p-b)*(p-c)
    S_real = math.sqrt(S_square)
    print("Dien tich tam giac:",round(S_real,1))
    print("Chu vi tam giac:", a+b+c)
else:
    print("Khong la 3 canh cua tam giac")
```

```python
#Bai 10
import math
Ox,Oy, R = map(float, input("Nhập toạ độ x,ytâm và bán kính R").split())
Ax, Ay = map(float, input("Nhập toạ độ x,y của A").split())
DodaidoanAO = math.sqrt((Ax-Ox)**2+(Ay-Oy)**2)
if DodaidoanAO <= R:
    print("A thuộc đường tròn")
else:
    print("Không thuộc đường tròn")
```

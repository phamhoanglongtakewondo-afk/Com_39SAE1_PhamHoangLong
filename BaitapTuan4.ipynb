```python
#W4A6
import math
a,b = map(int,input("Nhap 2 so a,b").split())
def songuyento(n):
    if n < 2:
        return False
    for e in range(2,int(math.sqrt(n))+1):
        if n%e ==0:
            return False
    return True
while a>b:
    a,b = map(int,input("Nhap 2 so a,b").split())
else:
    result = 0
    for i in range(a,b+1):
        if songuyento(i) is True:
            result += i
print(result)

```
